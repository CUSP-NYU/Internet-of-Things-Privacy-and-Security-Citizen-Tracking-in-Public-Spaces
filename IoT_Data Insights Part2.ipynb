## Company merged UUIDs insights

```python
company = pd.read_csv('/Users/smritimohta/Desktop/Capstone/Final/Code/Beacon_Manufacturers.csv')
company = company[['Company_Names','Example_UUID','Namespace_id']].rename(columns={'Namespace_id':'namespace_id'})
company.head()
```

```python
# processing the data
company['Example_UUID'] = company['Example_UUID'].str.lower()
company['uuid_code'] = company['Example_UUID'].str.slice(0,8)
company['uuid_code'][10] = company['namespace_id'][10]
company.head()
```

```python
# merging with main dataframe
df_merge = df.merge(company, on='uuid_code', how='left')
df_merge.head()
```

```python
# sorting to count numbers of company matched beacon products : ibeacon & altbeacon protocol - 13 uuids matched 
df_merge[['uuid','beacon_type','Company_Names']].groupby(by=['Company_Names',
                        'uuid',]).count().sort_values(by='beacon_type',ascending=False).reset_index()
```

```python
# sorting to count numbers of company matched beacon products : eddystone protocol - 0 uuids matched 
df_merge[['namespace_id_x','beacon_type','Company_Names']].groupby(by=['Company_Names',
                        'namespace_id_x',]).count().sort_values(by='beacon_type',ascending=False).reset_index()
```
