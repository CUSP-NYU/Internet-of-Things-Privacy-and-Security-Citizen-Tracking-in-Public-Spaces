```python
#data loading
df=pd.read_csv('/Users/anboguo/Desktop/Cap/Final/data_30sec.csv')
df=df.drop(columns='Unnamed: 0')
df['geom'] = df['latitude'].map(str) + ',' + df['longitude'].map(str)
df.head()
```

```python
#reverse geocoding
locator = Nominatim(user_agent='myGeocoder', timeout=10)
rgeocode = RateLimiter(locator.reverse, min_delay_seconds=0.000001)
tqdm.pandas()
df['address'] = df['geom'].progress_apply(rgeocode)
df.head()
```

```python
dfround=df[['uuid','major','minor','latitude','longitude','geom']]
dfround['latitude']=round(dfround['latitude'],4)
dfround['longitude']=round(dfround['longitude'],4)
dfround['geom']='('+dfround['geom']+')'
dfround.head()
```

```python
dfmerge=dfround.groupby(['uuid','major','minor','latitude','longitude']).count().reset_index()
geoms=[]
for i in range(len(dfmerge)):
    geoms.append(' '.join(dfround[(dfround['uuid']==dfmerge['uuid'][i]) & (dfround['major']==dfmerge['major'][i]) & (dfround['minor']==dfmerge['minor'][i]) & (dfround['latitude']==dfmerge['latitude'][i]) & (dfround['longitude']==dfmerge['longitude'][i])]['geom']))
dfmerge['geoms']=geoms
info=[]
for j in range(len(dfmerge)):
    info.append('uuid: '+' '+dfmerge['uuid'][j]+' '+'major: '+str(dfmerge['major'][j])+' '+'minor: '+str(dfmerge['minor'][j])+' '+'location: '+' '+dfmerge['geoms'][j])
dfmerge['info']=info
dfmerge.head()
```

```python
dfr=dfmerge[['latitude','longitude','info']]
dfr=dfr.drop_duplicates()
dfr.head()
```

```python
#add street 
data=pd.read_csv('/Users/anboguo/Desktop/Cap/Final/final_merged_data.csv')
data=data[['latitude','longitude','display_name']]
data['latitude']=round(data['latitude'],4)
data['longitude']=round(data['longitude'],4)
data.head()
```
