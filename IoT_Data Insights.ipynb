## Company merge

```python
company = pd.read_csv('/Users/smritimohta/Desktop/Capstone/Final/Code/Beacon_Manufacturers.csv')
company = company[['Company_Names','Example_UUID','Namespace_id']].rename(columns={'Namespace_id':'namespace_id'})
company.head()
```

```python
df_merge = df.merge(company, on='uuid_code', how='left')
df_merge.head()
```

## Conversion to geodataframe

```python
import geopandas as gpd
df_gpd = gpd.GeoDataFrame(df_merge, geometry=gpd.points_from_xy(df_merge.longitude, df_merge.latitude))
df_gpd = df_gpd.drop(columns=['Example_UUID','namespace_id_y'])
df_gpd.shape
```

## Reverse Geocoding

```python
import geopandas as gpd
import geopy
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
```

```python
lats=df_gpd['latitude'].to_list()
lons=df_gpd['longitude'].to_list()
# Creating a zip with latitudes and longitudes
coords=list(zip(lats,lons))
```

```python
len(coords)
```

```python
geolocator = Nominatim(user_agent="test_app")
location = geolocator.reverse("31.2158271,-85.3634326")
location.raw
```

```python
full_address=[]
for i in range(len(coords)):
    location = geolocator.reverse(coords[i])
    address=location.raw
    full_address.append(address)
```

```python
full_address 
```

```python
full_address
```

```python
full_address_df = pd.DataFrame(full_address)
full_address_df.head()
```

```python
full_address_df.to_csv('/Users/smritimohta/Desktop/Capstone/Final/Code/addresses.csv')
```

```python
address = full_address_df["address"].apply(pd.Series)
address.head()
```

```python
df_final = pd.concat([df_gpd, full_address_df, address], axis=1)
df_final.head()
```

```python
df_final = df_final.drop(columns=['Unnamed: 0','address'])
df_final.head()
```

## Counts

```python
ibeacon = df_final[df_final['beacon_type']=='ibeacon']
ibeacon = ibeacon[['beacon_type','uuid','census_id']].groupby(by=['uuid','census_id']).count().rename(columns={'beacon_type':'count'}).reset_index()
ibeacon = ibeacon.sort_values(by='count',ascending=False)
ibeacon
```

```python
altbeacon = df_final[df_final['beacon_type']=='altbeacon']
altbeacon = altbeacon[['beacon_type','uuid','census_id']].groupby(by=['uuid','census_id']).count().rename(columns={'beacon_type':'count'}).reset_index()
altbeacon = altbeacon.sort_values(by='count',ascending=False)
altbeacon
```

```python
eddybeacon = df_final[df_final['beacon_type']=='eddystone_uid']
eddybeacon = eddybeacon[['beacon_type','namespace_id_x','census_id']].groupby(by=['namespace_id_x','census_id']).count().rename(columns={'beacon_type':'count'}).reset_index()
eddybeacon = eddybeacon.sort_values(by='count',ascending=False)
eddybeacon = eddybeacon.rename(columns={'namespace_id_x':'uuid'})
eddybeacon
```
